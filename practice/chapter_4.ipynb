### Aggregation and Pivot Tables

```python
# Number of neighborhoods that bound the Charles River vs those that do not

bostonHousing_df.CHAS.value_counts()
```

```python
# code for aggregating MEDV by CHAS and RM

# create bins(구간) of size 1 for variables using the mathod pd.cut. 
# By default, the method creates a categorical variables. e.g. (6,7]. 
# The argument labels = False determines integers instead, e.g. 6.

bostonHousing_df['RM_bin'] = pd.cut(bostonHousing_df.RM, range(0,10), labels = False)
bostonHousing_df.head()
```

```python
# Compute the average of MEDV by (binned) RM and CHAS. 
# First group the data frame using the groupby method,
# then restrict(제한한다) the analysis to MEDV and determine the mean for each group

bostonHousing_df.groupby(['RM_bin', "CHAS"]).MEDV.mean()
```

```python
# code for creating pivot tables using the method pivot_table()

bostonHousing_df = pd.read_csv('BostonHousing.csv')
bostonHousing_df = bostonHousing_df.rename(columns = {'CAT. MEDV' : 'CAT_MEDV'})
# create bins of size 1
bostonHousing_df['RM_bins'] = pd.cut(bostonHousing_df.RM, range(0,10), labels = False)
```

```python
# use pivot_table() to reshape data and generate pivot table

pd.pivot_table(bostonHousing_df, values = 'MEDV', index=['RM_bins'],columns=['CHAS'],
              aggfunc = np.mean, margins = True)
```

## 4.6 Reducting the Number of Categories in Categorical Variables

```python
# code for creating Figure 4.1
```

```python
# use method crosstab to create a cross-tabulation of two variables
tb1 = pd.crosstab(bostonHousing_df.CAT_MEDV, bostonHousing_df.ZN)
tb1
```

```python
# convert numbers to ratios
propTb1 = tb1 / tb1.sum()
propTb1.round(2)
```

```python
# plot the ratios in a stacked bar chart
ax = propTb1.transpose().plot(kind='bar', stacked = True, width=0.9) 
#transpose() : 행과 열을 바꿔줌 (4,2) -> (2,4) / stacked = True : 0과 1의 막대가 동시에 쌓여서(합쳐서) 나옴
ax.set_yticklabels('{:.0%}'.format(x) for x in ax.get_yticks())
plt.title('Distribution of CAT.MEDV by ZN')
plt.legend(loc = 'center left', bbox_to_anchor=(1,0.5))
plt.show()
```

## 4.8 Principal Components Aanlysis (주성분분석, PCA)

참고(글) : https://angeloyeo.github.io/2019/07/27/PCA.html, https://m.blog.naver.com/tjdrud1323/221720259834

참고(영상) : https://www.youtube.com/watch?v=YEdscCNsinU

‘데이터 벡터를 어떤 벡터에 내적(혹은 정사영)하는 것이 최적의 결과를 내주는가?’ 이다. (='잘 요약하는' 법)

이 문제에 대한 해결은 바로 **공분산 행렬의 고윳값과 고유 벡터를 구함으로써** 가능하다  <-  **PCA**

- 고윳값 : 가중치
- 고유 벡터 : 정사영 당할 벡터

```python
# code for running PCA
```

```python
cereals_df = pd.read_csv("Cereals.csv")
cereals_df.head()
```

```python
pcs = PCA(n_components = 2) # components 개수 : 1개
pcs.fit(cereals_df[['calories', 'rating']])
```

```python
pcsSummary = pd.DataFrame({'Standard deviation':np.sqrt(pcs.explained_variance_), 
                         'Proportion of variance' : pcs.explained_variance_ratio_, 
                         'Cumulative proportion' : np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary
```

```python
pcsSummary = pcsSummary.transpose()
pcsSummary
```

```python
pcsSummary.columns = ['PC1','PC2'] # PC(principal components)1, PC2 : 주성분 2걔 (직선 그거)
pcsSummary.round(4)
```

```python
pcs.components_
# Z1의 가중치 : (-0.847, 0.532), Z2의 가중치 : (0.523, 0.847) <- 가중치 = 고유값
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns=['PC1','PC2'], index = ['calories','rating'])
pcsComponents_df
```

```python
scores = pd.DataFrame(pcs.transform(cereals_df[['calories','rating']]), columns = ['PC1','PC2'])
# caloires와 rating이 주성분 벡터에 정사영시킨 결과
scores.head()
```

### code for running PCA on the wine data

```python
pcs = PCA()
pcs.fit(cereals_df.iloc[:, 3:].dropna(axis=0))
pcsSummary_df = pd.DataFrame({'Standard deviation' : np.sqrt(pcs.explained_variance_),
                             'Proportion of variance' : pcs.explained_variance_ratio_,
                             'Cumulative proportion' : np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns = pcsSummary_df.columns,
                               index = cereals_df.iloc[:,3:].columns)
pcsComponents_df.iloc[:,:5]
```

### code for running PCA on the wine data

```python
wine_df = pd.read_csv('Wine.csv')
wine_df = wine_df.drop(columns = ['Type'])
pcs = PCA()
pcs.fit(wine_df.dropna(axis=0))
pcsSummary_df = pd.DataFrame({'Standard Deviation' : np.sqrt(pcs.explained_variance_),
                              'Proportion of variance' : pcs.explained_variance_ratio_,
                              'Cumulative proportion' : np.cumsum(pcs.explained_variance_ratio_)
                             })
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns = pcsSummary_df.columns, 
                                index = wine_df.columns)
pcsComponents_df.iloc[:,:5]
```
