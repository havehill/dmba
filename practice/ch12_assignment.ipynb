```python
bank = pd.read_csv("UniversalBank.csv")
```

```python
bank = bank.drop(columns=['ZIP Code'])
bank = pd.get_dummies(bank, columns = ['Education','Family'])
train_df, valid_df = train_test_split(bank, test_size=0.4, random_state=1)
train_df.head()
```

```python
model = LinearDiscriminantAnalysis()
model.fit(train_df.drop(columns=['Personal Loan']), train_df['Personal Loan'])
```

```python
for (i,j) in zip(train_df.columns, range(0,17)):
    globals()['c_{}'.format(i)] = model.coef_[0, j]
```
