```python
univ = pd.read_csv("Universities.csv")
univ.head()
```

```python
univ.info()
```

```python
univ_dr = univ.dropna(axis=0)
univ_dr.info()
```

### b

```python
univ_ind = univ_dr.set_index('College Name')
univ_ind
univ_conti = univ_ind.drop(columns = ['State'])
```

```python
# 정규화
univ_norm = univ_conti.apply(preprocessing.scale, axis=0)
```

```python
# 유클리드 거리
d_norm = pairwise.pairwise_distances(univ_norm, metric = 'euclidean')
```

```python
# 완전 연결법
Z = linkage(univ_norm, method = 'complete') 
memb = fcluster(Z, 6, criterion = 'maxclust')
memb = pd.Series(memb, index = univ_norm.index)
memb
```

```python
df = univ_norm.copy()

df['clust'] = memb.values
df
```

```python
df_agg_1 = df[df['clust']==1].agg(['min','max','mean','std','median'])
df_agg_1
```

# 15-2

### a

```python
pha = pd.read_csv("Pharmaceuticals.csv")
pha.set_index('Name', inplace = True)
pha.head()
```

```python
pha = pha.drop(columns=['Symbol'])
print(len(pha))
print(len(pha.columns))
```

```python
pha_n = pha.iloc[:,0:9]
pha_n.head()
```

### b

```python
pha_norm = (pha_n - pha_n.mean())/pha_n.std()
Z = linkage(pha_norm, method='average')
fig = plt.figure(figsize=(15,6))
fig.subplots_adjust(bottom=0.23)
dendrogram(Z, labels = pha_norm.index, color_threshold=3.75)
plt.axhline(y=3.75, color='black',linewidth=0.5)
plt.show()
```

```python
memb = fcluster(linkage(pha_norm, 'average'),6, criterion='maxclust')
memb = pd.Series(memb, index=pha_norm.index)
for key, item in memb.groupby(memb):
    print(key, ': ', ', '.join(item.index))
```

```python
pha_norm.index = ['{}: {}'.format(cluster, state) for cluster, state in zip(memb, pha_norm.index)]
sns.clustermap(pha_norm, method='average',col_cluster=False, cmap='mako_r')
plt.show()
```

# 15.3

### a

```python
cereal = pd.read_csv("Cereals.csv")
cereal.head()
```

```python
cereal_dr = cereal.dropna(axis=0)
```

### a

```python
cereal_cont = cereal_dr.drop(columns=['name','mfr','type'])
cereal_norm = cereal_cont.apply(preprocessing.scale, axis=0)
```

```python
Z = linkage(cereal_norm, method = 'single') 
fig = plt.figure(figsize=(15,6))
fig.subplots_adjust(bottom=0.23)
dendrogram(Z, labels = cereal_norm.index, color_threshold=2.7)
plt.axhline(y=2.7, color='black',linewidth=0.5)
plt.show()
```

```python
Z = linkage(cereal_norm, method = 'complete') 
fig = plt.figure(figsize=(15,6))
fig.subplots_adjust(bottom=0.23)
dendrogram(Z, labels = cereal_norm.index, color_threshold=5.7)
plt.axhline(y=5.7, color='black',linewidth=0.5)
plt.show()
```
